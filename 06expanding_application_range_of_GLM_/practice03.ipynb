```python
def logistic(beta1,beta2,x):
    return 1 / (1+np.exp(-(beta1+beta2*x)))

x = np.arange(-3,4,0.1)
plt.plot(x,logistic(0,2,x),label='$\\beta_{1} = 0$')
plt.plot(x,logistic(2,2,x),label='$\\beta_{1}=2$')
plt.plot(x,logistic(-3,2,x),label='$\\beta_{1}=-3$')
plt.legend(loc='upper left')
```

```python
plt.title('$\\beta_{1}=0$',fontsize=15)
plt.plot(x,logistic(0,-1,x),label='$\\beta_{2}=-1$')
plt.plot(x,logistic(0,2,x),label='$\\beta_{2}=2$')
plt.plot(x,logistic(0,4,x),label='$\\beta_{2}=4$')
plt.legend(loc='center left')
```

```python
data['f_int'] = (data.f=='T').astype(int)
```

```python
data['N-y'] = data['N'] - data['y']
```

```python
data.x_c = sm.add_constant(data[['x','f_int']])
model = sm.GLM(data[['y','N-y']],data.x_c,family=sm.families.Binomial())
result = model.fit()
result.summary()
```

```python
data['predict'] = result.predict()
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
sort_data = data.sort_values('x')
ax[0].scatter(data[data.f=='T'].x,data[data.f=='T'].y)
ax[0].plot(sort_data[data.f=='T'].x,sort_data[data.f=='T'].predict*8)
ax[0].set(xlabel='$x_{i}$',ylabel='$y_{i}$',title='$f_{i}=T$')

ax[1].scatter(data[data.f=='C'].x,data[data.f=='C'].y,facecolor='none')
ax[1].plot(sort_data[data.f=='C'].x,sort_data[data.f=='C'].predict*8)
```

```python
summary = []
llh = np.sum(stats.binom.logpmf(data.y,8,data.y/8.))
k = 100
deviance = -2 * llh
residual_deviance = 0
aic = -2 * (llh - k)
summary.append(['full',k,llh,deviance,residual_deviance,aic])
```

```python
model = sm.GLM(data[['y','N-y']],data.x_c['const'],family=sm.families.Binomial())
result = model.fit()
summary.append(['const',result.df_model+1,result.llf,-2*result.llf,result.deviance,result.aic])
```

```python
import itertools
columns = ['x','f_int']
for i in range(2):
    for a in itertools.combinations(columns,i+1):
        data.x_c = sm.add_constant(data[list(a)])
        model = sm.GLM(data[['y','N-y']],data.x_c,family=sm.families.Binomial())
        result = model.fit()
        summary.append([list(a),result.df_model+1,result.llf,-2*result.llf,result.deviance,result.aic])
```

```python
summary_df = pd.DataFrame(summary,columns=['model','degree of freedom','log likelihood','deviance','residual deviance','aic'])
```
