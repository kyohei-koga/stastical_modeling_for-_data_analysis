```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

data = pd.read_csv('data3a.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data[data.f=='C'].x,data[data.f=='C'].y,facecolor='none',label='C')
plt.scatter(data[data.f=='T'].x,data[data.f=='T'].y,label='T')
plt.xlabel('size')
plt.ylabel('num of seed')
```

```python
data.boxplot('y',by='f')
```

```python
x = np.arange(-4,4,0.01)
plt.plot(x,np.exp(-1+0.4*x),label='{-1,0.4}')
plt.plot(x,np.exp(-2-0.8*x),label='{-2,-0.8}')
plt.legend(loc='upper right',title='{B1,B2}')
plt.xlabel('body size x_i of i')
plt.ylabel('lambda_i of i')
```

```python
import statsmodels.api as sm
```

```python
data.x_c = sm.add_constant(data.x)
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
```

```python
result = model.fit()
result.summary()
```

```python
plt.scatter(data.x[data.f=='C'],data.y[data.f=='C'],label='C',facecolors='none')
plt.scatter(data.x[data.f=='T'],data.y[data.f=='T'],label='T')

x = np.arange(min(data.x),max(data.x),(max(data.x)-min(data.x))/100.)
plt.plot(data.sort_index(by='x').x,result.predict(data.x_c.sort_index(by='x')))
```

```python
data['f_int'] = (data.f=='T').astype(int)
```

```python
data.x_c = sm.add_constant(data['f_int'])
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
```

```python
model.fit().summary()
```

```python
data.x_c = sm.add_constant(data[['x','f_int']])
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
```

```python
model.fit().summary()
```
