```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels
from scipy import stats
```

```python
data = pd.read_csv('data3a.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data[data.f=='T'].x,data[data.f=='T'].y,facecolor='none',label='T')
plt.scatter(data[data.f=='C'].x,data[data.f=='C'].y,label='C')
plt.legend(loc='upper left')
```

```python
data.boxplot('y',by='f')
```

```python
x = np.arange(-4,4,0.01)

plt.plot(x,np.exp(-1+0.4*x),label='{-1,0.4}')
plt.plot(x,np.exp(-2-0.8*x),label='{-2,-0.8}')
plt.legend(loc='upper right',title='{B1,B2}')
```

```python
data.x_c = sm.add_constant(data.x)
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
plt.scatter(data.x,data.y,label='observation')
plt.scatter(data.x_c['x'],result.predict(data.x_c),c='r',alpha=0.5)
```

```python
data['f_int'] = (data.f=='T').astype(int)
data.x_c = sm.add_constant(data.f_int)
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
data.x_c = sm.add_constant(data[['x','f_int']])
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
```

```python
result = model.fit()
result.summary()
```
