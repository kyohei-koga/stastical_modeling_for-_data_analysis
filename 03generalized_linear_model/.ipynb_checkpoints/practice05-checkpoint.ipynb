```python
%matplotlib inline
import pandas as pd
import numpy as np
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('data3a.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data[data.f=='T'].x,data[data.f=='T'].y,label='T')
plt.scatter(data[data.f=='C'].x,data[data.f=='C'].y,label='C',facecolor='none')
plt.xlabel('size')
plt.ylabel('num of seed')
plt.legend()
```

```python
data.boxplot('y',by='f')
```

```python
x = np.arange(-4,4,0.1)
plt.plot(x,np.exp(-1+0.4*x),label='{-1,0.4}')
plt.plot(x,np.exp(-2-0.8*x),label='{-2,-0.8}')
plt.xlabel('body sixe x_i of i')
plt.ylabel('lambda_i of i')
plt.legend(title='{B1,B2}')
```

```python
data.x_c = sm.add_constant(data.x)
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
x_c = data.x_c.sort_index(by='x')
plt.plot(x_c['x'],result.predict(x_c))
plt.scatter(data[data.f=='C'].x,data[data.f=='C'].y,facecolor='none')
plt.scatter(data[data.f=='T'].x,data[data.f=='T'].y)
```

```python
data['f_int'] = (data.f=='T').astype(int)
```

```python
data.x_c = sm.add_constant(data.f_int)
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
data.x_c = sm.add_constant(data[['x','f_int']])
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
beta1,beta2,beta3 = result.params
```

```python
print(beta1,beta2,beta3)
```

```python
x = np.arange(5,20,0.1)
plt.plot(x,np.exp(beta1+beta2*x+beta3),label='T')
plt.plot(x,np.exp(beta1+beta2*x),label='C')
plt.legend(loc='upper left')
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson(statsmodels.genmod.families.links.identity))
```

```python
result = model.fit()
result.summary()
```
