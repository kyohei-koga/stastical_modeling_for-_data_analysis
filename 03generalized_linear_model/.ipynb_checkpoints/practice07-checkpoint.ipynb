```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data3a.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data.x[data.f=='C'],data.y[data.f=='C'],facecolor='none',label='C')
plt.scatter(data.x[data.f=='T'],data.y[data.f=='T'],label='T')
plt.legend(loc='upper left')
```

```python
data.boxplot('y',by='f')
```

```python
import numpy as np
x = np.arange(-4,4,0.01)

plt.plot(x,np.exp(-1+0.4*x),label='{-1,0.4}')
plt.plot(x,np.exp(-2-0.8*x),label='{-2,-0.8}')
plt.legend(title='{B1,B2}')
```

```python
import statsmodels.api as sm
```

```python
data.x_c = sm.add_constant(data.x)
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
plt.scatter(data.x[data.f=='C'],data.y[data.f=='C'],facecolor='none')
plt.scatter(data.x[data.f=='T'],data.y[data.f=='T'])

plt.plot(data.sort_index(by='x').x,result.predict(data.x_c.sort_index(by='x')),c='b',alpha=0.5)
```

```python
data['f_int'] = (data.f == 'T').astype(int)
data.x_c = sm.add_constant(data['f_int'])
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
data.x_c = sm.add_constant(data[['x','f_int']])
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
result.summary()
```

```python
import statsmodels
data.x_c = sm.add_constant(data[['x','f_int']])
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson(statsmodels.genmod.families.links.identity))
result = model.fit()
```

```python
result.summary()
```
