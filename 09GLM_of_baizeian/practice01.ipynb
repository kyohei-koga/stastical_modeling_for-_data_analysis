```python
%matplotlib inline
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data_T = pd.read_csv('d.csv')
```

```python
data = data_T.T[1:]
```

```python
data = data.rename(columns={0:'x',1:'y'})
```

```python
data.x_c = sm.add_constant(data.x)
```

```python
model = sm.GLM(data.y,data.x_c,family=sm.families.Poisson())
result = model.fit()
```

```python
result.params
```

```python
data['prediction'] =result.predict(data.x_c)
```

```python
plt.scatter(data.x,data.y,facecolor='none',label='sample')
plt.plot(data.sort_values('x').x,data.sort_values('x').prediction,label='population',ls='--')
plt.legend(loc='upper right')
```

```python
result.summary()
```

```python
y = np.linspace(-20,20,100)
plt.plot(y,stats.norm.pdf(y,0,100))
plt.plot(y,stats.norm.pdf(y,0,1),ls='--')
plt.show()
```

```python
import pymc
```

```python
basic_model = pymc.Model()

beta1 = pymc.Normal('beta1',0.,tau=1.0e-3)
beta2 = pymc.Normal('beta2',0.,tau=1.0e-3)

@pymc.deterministic(plot=False)
def mu(beta1=beta1,beta2=beta2,x=data.x):
    return np.exp(beta1+beta2*(x-x.mean()))

Y_obs = pymc.Poisson('Y_obs',mu=mu,value=data.y,observed=True)

model = pymc.Model([beta1,beta2,Y_obs])
start = pymc.MCMC(model)
start.sample(iter=105000,burn=5000,thin=100)
```

```python
pymc.Matplot.plot(start)
```

```python
import pydot
import scipy.misc
from PIL import Image
```

```python
pymc.graph.graph(model,format='png',path='',name='model',prog='dot')
img = np.array(Image.open('model.png'))
plt.imshow(img)
```
