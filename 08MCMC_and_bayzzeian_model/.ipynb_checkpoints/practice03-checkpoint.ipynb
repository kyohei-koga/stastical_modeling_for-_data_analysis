```python
def series_llf(q,steps=100):
    series = [[0,q]]
    llh = np.sum(stats.binom.logpmf(data,8,q))
    for i in range(steps):
        random = np.random.randint(2)
        if random == 1:
            cal = np.sum(stats.binom.logpmf(data,8,q+0.01))
            if cal > llh:
                q += 0.01
                llh = cal
        else:
            cal = np.sum(stats.binom.logpmf(data,8,q-0.01))
            if cal > llh:
                q -= 0.01
                llh = cal
        series.append([i,q])
        
    return np.array(series)

init_3 = series_llf(0.3)
init_6 = series_llf(0.6)
```

```python
plt.plot(init_3[:,0],init_3[:,1],'--',label=0.3)
plt.plot(init_6[:,0],init_6[:,1],'--',label=0.6)
plt.legend()
```

```python
def series_llf_metro(prob,size):
    llf = np.sum(stats.binom.logpmf(data,8,prob))
    series = [[0,prob,llf]]
    for i in range(size):
        random = np.random.randint(2)
        if random == 1:
            cal = np.sum(stats.binom.logpmf(data,8,prob+0.01))
            if cal > llf:
                llf = cal
                prob += 0.01
            elif np.random.rand() < np.exp(cal-llf):
                llf = cal
                prob += 0.01
        else:
            cal = np.sum(stats.binom.logpmf(data,8,prob-0.01))
            if cal > llf:
                llf = cal
                prob -= 0.01
            elif np.random.rand() < np.exp(cal-llf):
                llf = cal
                prob -= 0.01
        series.append([i+1,prob,llf])
        
    return np.array(series)

init_3_100 = series_llf_metro(0.3,100)
init_3_1000 = series_llf_metro(0.3,1000)
init_3_10000 = series_llf_metro(0.3,10000)
```

```python
fig,ax = plt.subplots(3,2,figsize=(20,15))
steps = [init_3_100,init_3_1000,init_3_10000]
for i,step in enumerate(steps):
    #print step
    ax[i][0].plot(step[:,0],step[:,1])
    ax[i][0].set_ylim(0.2,0.65)
    ax[i][1].hist(step[:,1],orientation='horizontal',bins=np.arange(0.245,0.655,0.01))
    
```
