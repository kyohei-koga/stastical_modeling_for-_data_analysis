```python
x = np.arange(0,20)
plt.plot(pd.Series(stats.poisson.pmf(x,3.5),index=x),'o--',label=3.5)
plt.plot(pd.Series(stats.poisson.pmf(x,7.1),index=x),'D--',label=7.1)
plt.plot(pd.Series(stats.poisson.pmf(x,15.1),index=x),'^--',label=15.1)
plt.xlabel('x')
plt.ylabel('prob')
plt.legend(loc='upper right',title='lambda')
```

```python
t_lambda = [2.0,2.4,2.8,3.2,3.6,4.0,4.4,4.8,5.2]
```

```python
pieces = []
x = np.arange(0,10)
for lam in t_lambda:
    series = pd.Series(stats.poisson.pmf(x,lam),index=x,name='lambda_%.1f'%lam)*50
    pieces.append(series)
    
dd = pd.concat(pieces,axis=1)
```

```python
dd.head()
```

```python
import itertools
fig, axes = plt.subplots(3,3,figsize=(15,15))
for i,j in enumerate(itertools.product(range(3),range(3))):
    dd['lambda_%.1f'%t_lambda[i]].plot(style='o--',label='lambda%.1f'%t_lambda[i],ax=axes[j[0]][j[1]])
    loglikelihood = np.sum(stats.poisson.logpmf(data,t_lambda[i]))
    axes[j[0]][j[1]].hist(data,bins=np.arange(-0.5,8.5,1.),label='log L=%.1f'%loglikelihood)
    axes[j[0]][j[1]].legend(loc='upper right')
    axes[j[0]][j[1]].set_yticks((0,14))
    axes[j[0]][j[1]].set_xticks((-2,10))
```

```python
x = np.arange(2.0,5.0,0.01)
logL = np.array([np.sum(stats.poisson.logpmf(data,i)) for i in x])
plt.plot(x,logL)
plt.axvline(x[np.argmax(logL)],c='r')
plt.xlabel('lambda')
plt.ylabel('logL')
```

```python
print(x[np.argmax(logL)])
```

```python
likeliwoods = []
for j in range(3000):
    data = np.random.poisson(3.5,50)
    x = np.arange(2.0,5.0,0.01)
    logL = np.array([np.sum(stats.poisson.logpmf(data,i)) for i in x])
    likeliwoods.append(x[np.argmax(logL)])
```

```python
plt.hist(likeliwoods,bins=np.arange(2.0,5.0,0.1))
plt.xlabel('value of likeliwoods')
plt.ylabel('num of likeliwood')
```
