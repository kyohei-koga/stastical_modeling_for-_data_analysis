# 階層ベイズメールーGLMMのベイズモデル化ー

階層事前分布を使って、一般化線形混合モデルを階層ベイズモデルとして扱う方法を説明します。

統計モデルを複雑にすると、パラメータ推定も難しくなります。階層ベイズモデルとMCMCサンプリングによるパラメーター推定という組み合わせが威力を発揮するのはこのような状況です。

## 10.2GLMMの階層ベイズモデル化
個体ごとに由来する原因不明な差異などを組み込んだGLMがGLMMです。ここではGLMMのベイズモデル化に取り組みます。

ここで登場するGLMMではリンク関数と線形予測子を$logit(q_{i})=\beta + r_{i}$とする。切片$\beta$は全個体に共通するパラメータであり、個体差を表す$r_{i}$は平均ゼロでsの正規分布に従うと仮定します。

今回推定したい事後分布は以下のようになるので、あとは事前分布を指定すればモデルの設計は完了します。


事後分布$\propto p({\bf Y}|\beta,{r_{i}})\times$事前分布

$\beta$は特に制約はないので、無情報事前分布とする。

個体差$r_{i}$のパラメータの事前分布はどうすれば良いか。以前は、個体ごとに異なるパラメータ$r_{i}$が、いずれも平均ゼロで標準偏差sの正規分布従うと仮定していました。ここでも同じように$r_{i}$の事前分布が

$$p(r_{i}|s)=\frac{1}{\sqrt{2 \pi s^{2}}}exp(\frac{-r^{2}_{i}}{2s^{2}})$$

であるとする。最後に残された問題は、この個体差のばらつきsというパラメータの扱いです。ベイズ統計モデルではsの事後分布を推定します。

事後分布を推定するためには事前分布指定しなければなりません。sは正のあたいであれば何だったいいので無情報事前分布としてよい。ここでは一様分布を使う。

sはハイパーパラメータと呼ばれることもある。

## 10.3階層ベイズモデルの推定・予測
ベイズモデルでは、「推定したいパラメータ」はことごとく事前分布とデータに基づいた事後分布が推定されます。つまり、この例題の階層ベイズモデルの事後分布とは、
$$p(\beta,s,r_{i}|{\bf Y}) \propto p({\bf Y}|\beta,r_{i})p(\beta)p(s)\Pi_{i}p(r_{i}|s)$$

```python
Y = data.y
N = len(Y)
s = pymc.Uniform(name='s',lower=0,upper = 1.0e+4,value=0.01) #階層事前分布のパラメータの事前分布
b = pymc.Normal(name='b', mu=0.0, tau=1.0e+2) #ベータの事前分布

r = [pymc.Normal(name='r_{0}'.format(i),mu=0,tau=s**(-2)) for i in range(N)] #階層事前分布
p = pymc.Lambda('p',lambda r = r,b=b:pymc.invlogit(b+r)) #q=1/(1+exp(-(beta+r)))

obs = pymc.Binomial(name='obs',n=8,p=p,value=Y,observed=True)
model = pymc.Model([b,s,obs])
```

```python
mcmc = pymc.MCMC(model)
mcmc.sample(iter=10000,burn=1000,thin=10)
```

```python
plt.figure(figsize=(100,10))
pymc.graph.graph(model, format='png', path='',name='model',prog='dot')
img = np.array(Image.open('model.png'))
plt.imshow(img)
```

### 10.3.2 階層ベイズモデルの事後分布推定と予測
生存種子数yの確率分布は、二項分布$p(y|\beta,r)$と正規分布$p(r|s)$の無限混合分布であり、以下のような式で表せるとする。
$$p(y|\beta,s)=\int_{-\infty}^{\infty}p(y|\beta,r)p(r|s)dr$$

ここで登場するrについての積分とは、事後分布p(r|s)に従うような事物を無限個集めてきて、その平均を評価しているという意味だと考えてください。

```python
pymc.Matplot.plot(mcmc.trace('b'),common_scale=False)
```

```python
pymc.Matplot.plot(mcmc.trace('s'),common_scale=False)
```

```python
pymc.Matplot.plot(mcmc.trace('r_0'),common_scale=False)
```

```python
pymc.Matplot.plot(mcmc.trace('r_1'),common_scale=False)
```

```python
pymc.Matplot.plot(mcmc.trace('r_2'),common_scale=False)
```

600あたりくらいからうまくサンプリングできていいそう。

```python
left = 600

L = []
b_= np.mean(mcmc.trace('b')[left:])
#平均値を計算
for i in range(N):
    ri = np.mean(mcmc.trace('r_{0}'.format(i))[left:])
    L.append(pymc.binomial_expval(n=8,p=pymc.invlogit(b_+ri)))
```

```python
n_b, bins_b, pathces_b = plt.hist(data.y,bins=np.linspace(-0.5,8.5,10),color='white',label='before')
n_a, bins_a, pathces_a = plt.hist(L,bins=np.linspace(-0.5,8.5,10),color='blue',label='after',alpha=0.3)
plt.legend()
plt.show()
```

推定された事後分布を組み合わせて、生存種子数ごとの個体数の分布を予測

```python
plt.plot(np.linspace(0,8,9),n_b,'ro--',label='before')
plt.plot(np.linspace(0,8,9),n_a,'bo',label='after',alpha=0.5)
plt.legend(loc='upper center')
```

```python
left = 600
Big = []
for i in range(mcmc.trace('b')[left:].shape[0]):
    data = []
    for j in range(N):
        r = mcmc.trace('r_{0}'.format(j))[left+i]
        q = pymc.binomial_expval(n=8,p=pymc.invlogit(mcmc.trace('b')[left+i]+r))
        data.append(q)
    Big.append(data)
```

```python
count_data = np.round(Big)
```

```python
count_df = []
for i in range(count_data.shape[0]):
    count = np.bincount(count_data.astype(int)[i],minlength=9)
    count_df.append(count)
```

```python
df = pd.DataFrame(count_df)
```

```python
percentile_low = np.percentile(df,2.5,axis=0)
percentile_high = np.percentile(df,97.5,axis=0)
percentile_mid = np.percentile(df,50,axis=0)
```

```python
plt.fill_between(np.linspace(0,8,9),percentile_high,percentile_low,alpha=0.3)
plt.plot(np.linspace(0,8,9),n_b,'ro',label='before')
plt.plot(np.linspace(0,8,9),percentile_mid,'bo--',label='after',alpha=0.5)
plt.xlabel('num of survived species:y')
plt.ylabel('num of plant')
plt.legend(loc='upper center')
plt.show()
```

## 10.4ベイズモデルで使う様々な事前分布
ベイズモデルの設計で重要になるのは、事前分布の選択です。

あるパラメータの事前分布として、無情報事前分布・階層事前分布のどちらを選択すべきなのでしょうか。その選択はそのパラメータがデータ全体のどの範囲を説明しているかに依存すると考えて良い。

データ全体を大域的に説明する少数のパラメータと、データのごく一部だけを説明する局所的な多数のパラメータがあります。

$r_{i}$の事前分布として無情報事前分布を指定たとしましょう。このようなモデリングでは、100個の$r_{i}$それぞれを自由に決めるので、統計モデルとして望ましいものではありません。これは「１番目の個体はこうなっていて、２番目は・・・」と「データの読み上げ」みたいなもので、統計モデルではありません。

ここでは個体差$r_{i}$の事前分布$p(r_{i}|s)$を階層事前分布として設定し、$p(r_{i}|s)$が平均ゼロで標準偏差sの正規分布であるとしました。そして、このsの事前分布$p(s)$は無情報事前分布であると設定したので、階層はこれ以上深くなりません。統計モデルをこのように設計しておくと、他のパラメータと同時にsの事後分布もデータにうまくあてはまるように推定されます。

あるパラメータの事前分布として、事前情報分布・階層事前分布のどちらを選ぶべきかは、そのパラメータがデータ全体のどの範囲を説明しているかに依存すると考えていよいでしょう。現実的な統計モデルには２種類のパラメータが含まれています。

- データ全体を大域的に説明する少数のパラメータ
- データのごく一部だけを説明する局所的な多数のパラメータです。

この例題で言うと、切片$\beta$は大域的なパラメータなので、無情報事前分布を用いています。一方、100個もある個体差${r_{i}}$は
- 個々の$r_{i}$は全体のごく一部を説明しているだけ
- 全個体の${r_{i}}$は「似たような（ある範囲に分布する）」パラメータの集まりと考えられる

と言った性質を持つパラメータである。個々の$r_{i}$に無情報事前分布を指定するのではなく、$r_{i}$全体のばらつきを変えられる階層事前分布を指定します。

階層ベイズモデルでは多数の局所的パラメータを階層事前分布で「拘束」して自由な当てはめを制約し、すべての局所的なパラメータの事後分布を推定できるようにします。


## 10.5 個体差+場所差の階層ベイズモデル
個体差だけでなく、場所差も組み込んだ階層ベイズモデルについて。

ここで知りたいのは、肥料をやる処理をした時の種子数$y_{i}$の変化です。

この実験では、植木鉢が10個準備されていて、各植木鉢に10個体の植物が育てられているとする。

```python
data = pd.read_csv('d1.csv')
data.columns
```

```python
data['pot_int'] = data['pot'].map(dict(zip(data['pot'].unique(),range(len(data['pot'].unique())))))
```

```python
data['f_int'] = (data['f']=='T').astype(int)
```

```python
data[data.f=='C'].describe()
```

無処理のyの平均は6.6程度、yがポアソン分布に従うなら標準偏差は2.6程度になるはずだが、観測値では、7.2程度で過分散であることがわかる。

```python
plt.scatter(data['id'],data['y'],c=data.pot_int,cmap=cm.Accent)
plt.colorbar()
```

```python
data.boxplot('y',by='pot')
plt.show()
```

過分散の原因は個体差だけでなく、植木鉢の差にもありそうであることがわかる。例えば、C,F,H,Jの植木鉢ではどの個体も種子数が少なくなっています。

このようなデータ構造の場合、個体も植木鉢もどちらも反復ではなく擬似反復なので、個体差と植木鉢差を同時に扱う統計モデルを作らなくては行けない。

これはGLMM化したポアソン回帰で扱える問題なので、これまでのように個体iの種子数$y_{i}$のばらつきを平均$\lambda_{i}$のポソン分布で表現します。
$$p(y_{i}|\lambda_{i})=\frac{\lambda^{y_{i}}_{i}exp(-\lambda_{i})}{y_{i}!}$$

対数リンク関数を使って平均種子数は以下とします。
$$log\lambda_{i}=\beta_{1}+\beta_{2}f_{i}+r_{i}+r_{j(i)}$$

個体差を表す$r_{i}$は100個、植木鉢の差である$r_{j(i)}$は10個あり、それぞれのばらつきを$s$と$s_{p}$とする。

パラメータの事前分布は以下のようにする。

- 大域的なヘインパラメータである$\beta_{1}$と$\beta_{2}$は無情報事前分布、平均ゼロの押しつぶされた正規分布

- 大域的なバラつきパラメータである$s$と$s_{p}$は無情報事前分布、0から$10^{4}$の範囲をとる一様分布

- 局所的パラメーターである$r_{i}$と$r_{j(i)}$は階層事前分布、平均ゼロで標準偏差はそれぞれ$s$と$s_{p}$

```python
b1 = pymc.Normal('b1',mu=0,tau=1.0e+2)
b2 = pymc.Normal('b2',mu=0,tau=1.0e+2)
s  = pymc.Uniform('s',lower=0,upper=1.0e+4)
sp = pymc.Uniform('sp',lower=0,upper=1.0e+4)

rs = [pymc.Normal(name='rs_{0}'.format(i),mu=0,tau=s**(-2)) for i in range(len(data))]
rp = [pymc.Normal(name='rp_{0}'.format(i),mu=0,tau=sp**(-2)) for i in range(len(data['pot'].unique()))]

@pymc.deterministic
def p(b1=b1,b2=b2,s=s,sp=sp,rs=rs,rp=rp):
    L = []
    for i in range(len(data)):
        lambda_i = np.exp(b1 + b2*data.iloc[i]['f_int'] + rs[i] + rp[data.iloc[i]['pot_int']])
        L.append(lambda_i)
    return np.asarray(L)
```

```python
obs = pymc.Poisson('obs',mu=p,value=data['y'],observed=True)
model = pymc.Model([b1,b2,s,sp,obs])
```

```python
mcmc = pymc.MCMC(model)
mcmc.sample(iter=10000,burn=1000,thin=50)
```

#### 反復と擬似反復
データをどのようにとったのかー個体差・場所差をどのように統計モデルに組み込むのか方法は、これに依存します。要点は個体差、場所差が「見えてしまう」データのとり方をしているかどうかです。

例として、種子の生存確率を知るために、植物から種子を撮ってきてその生死を調べたとしましょう。ここでは、調査対象の植物たちは植木鉢で育てられているとします。

- 各個体でひとつだけ種子の生死を調べた

この場合、過分散も生じませんし、個体差のばらつきも推定できません。更に、一つの植木鉢に１個体がいるだけなので個体差と植木鉢差の区別がつきません。このようなデータのとり方を反復とい言います。
$$logit(q_{i})=\beta_{1}+\beta_{2}x_{i}$$

- 同じ個体から採取したN個(N>1)の種子の生死を調べた

「この個体では種子がやたら死んでいる」、「こっちの個体はおなじ$x_{i}$なのに生存率がやらたと高い」と言ったことがデータから読み取れる。この場合、個体差のばらつきを入れた統計モデルが使える。

このような場合、「個体から複数のデータをとる」という実験操作は反復ではなく、擬似反復というなります。「擬似」とは反復っぽく見えるけれど個体差などを打ち消す反復になっていないぐらいの意味です。ただし、この場合でも、植木鉢は反復になっていて、個体に差があるのか植木鉢に差があるのか区別できないので、植木鉢だけを分離して推定できません。
$$logit(q_{i})=\beta_{1}+\beta_{2}x_{i}+r_{i}$$

- 各個体でひとつだけ種子の生死を調べたが、一つの植木鉢に複数の個体が植えられている。

「こっちの植木鉢では死亡率が高い」と言ったことが推定可能です。このような場合は植木鉢差を$r_{j}$として統計モデルに埋め込む

$$logit(q_{i})=\beta_{1}+\beta_{2}x_{i}+r_{j}$$

- 同じ個体からN個の種子を採取し、一つの植木鉢に複数の個体が植えられている。

この場合は、個体も、植木鉢も擬似反復になっているのでそれぞれを埋め込んだ統計モデルになる。

$$logit(q_{i})=\beta_{1}+\beta_{2}x_{i}+r_{i}+r_{j}$$
